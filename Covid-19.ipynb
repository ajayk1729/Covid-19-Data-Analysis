```python
#Converting the column data_type

#1.Replace non-finite values with 0,

#2.Convert to integer data type in case of string values so that we can easily visulaize
df['Deaths / 100 Recovered']=df['Deaths / 100 Recovered'].replace([np.inf, np.nan], 0).astype(int)
df['Deaths / 100 Recovered']  
```

```python
#checking data after modifying the data
df
```

```python
#to check if there is changes done by above processing or not
changes = df_original['Deaths / 100 Recovered'] != df['Deaths / 100 Recovered']
if changes.any():
    print("Changes have been made to the column.")
else:
    print("No changes have been made to the column.")
```

```python
#compare US data and India data (confirmed and recovered)
df_compare=df[df['Country/Region'].isin(['US','India'])][['Country/Region','Confirmed','Recovered']]
df_compare = df_compare.set_index('Country/Region')
df_compare
```

```python
df_compare.plot.bar()
```

```python
df_new = df[['Country/Region', 'New cases', 'New deaths', 'New recovered']]
df_new
```

```python
Country_based_new_case = df_new.groupby('Country/Region')['New cases'].sum()
Country_based_new_case
```

```python
#Top 10 countries
Top_10_countries = Country_based_new_case.sort_values(ascending=False).head(10)
plt.bar(Top_10_countries.index, Top_10_countries.values)
plt.xlabel('Country')
plt.ylabel('New cases')
plt.title('Top 10 Countries by New cases')
plt.xticks(rotation=45)
plt.show()
```
